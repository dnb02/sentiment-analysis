```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn import svm
```

```python
random_seed = 10
test_size = 0.2
```

```python
labels = data.price
train1 = data.drop(['id', 'date', 'price'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(train1, labels, test_size=test_size, random_state=random_seed)
x = list(train1.columns)
```

```python
print('linear regression without normalizing')
linear_reg = LinearRegression()
linear_reg.fit(x_train,y_train)
acc = round(linear_reg.score(x_test,y_test), 3)
print('R^2 score {}'.format(acc))
```

```python
fig, ax = plt.subplots()
plt.barh(x, linear_reg.coef_)
ax.set_xlabel('Importance')
ax.set_ylabel('Features')
ax.set_title('Feature Importance')
plt.show()
```

```python
print('linear regression with normalizing')
linear_reg = LinearRegression(normalize=True)
linear_reg.fit(x_train,y_train)
acc = round(linear_reg.score(x_test,y_test), 3)
print('R^2 score {}'.format(acc))
```

```python
fig, ax = plt.subplots()
plt.barh(x, linear_reg.coef_)
ax.set_xlabel('Importance')
ax.set_ylabel('Features')
ax.set_title('Feature Importance')
plt.show()
```

```python
rf_reg = ensemble.RandomForestRegressor(n_estimators = 400, max_depth = 5, \
                                        min_samples_split = 2, random_state=random_seed)
rf_reg.fit(x_train, y_train)
acc = round(rf_reg.score(x_test,y_test), 3)
print('R^2 score {}'.format(acc))
```

```python
fig, ax = plt.subplots()
plt.barh(x, rf_reg.feature_importances_)
ax.set_xlabel('Importance')
ax.set_ylabel('Features')
ax.set_title('Feature Importance')
plt.show()
```

```python
gb_reg = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5,
                                            min_samples_split = 2, learning_rate = 0.1, loss = 'squared_error', random_state=random_seed)
gb_reg.fit(x_train, y_train)
acc = round(gb_reg.score(x_test,y_test), 3)
print('R^2 score {}'.format(acc))
```

```python
fig, ax = plt.subplots()
plt.barh(x, gb_reg.feature_importances_)
ax.set_xlabel('Importance')
ax.set_ylabel('Features')
ax.set_title('Feature Importance')
plt.show()
```
